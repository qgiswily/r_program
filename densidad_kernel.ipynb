```python
import numpy as np
import matplotlib.pyplot as plt

def epanechnikov_kernel(u):
  """Calcula el valor del Kernel de Epanechnikov para un valor u dado."""
  if abs(u) <= 1:
    return (3/4) * (1 - u**2)
  else:
    return 0

# Generar valores de u para el gráfico
u_values = np.linspace(-1.5, 1.5, 400)  

# Calcular los valores del kernel para cada u
kernel_values = [epanechnikov_kernel(u) for u in u_values]

# Graficar el Kernel de Epanechnikov
plt.plot(u_values, kernel_values)
plt.title("Función del Kernel de Epanechnikov")
plt.xlabel("u")
plt.ylabel("K(u)")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
data = np.array([2, 3, 4, 6, 8]) 

# Ancho de banda
h = 1.5

# Función del kernel gaussiano
def gaussian_kernel(x, xi, h):
  return (1 / (h * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - xi) / h)**2)

# Rango de valores x para graficar
x_grid = np.linspace(0, 10, 200)  

# Calcular las contribuciones individuales de cada punto de datos
kernel_values = np.zeros((len(data), len(x_grid)))
for i, xi in enumerate(data):
  kernel_values[i, :] = gaussian_kernel(x_grid, xi, h)

# Calcular la función de densidad de kernel completa
density_estimate = np.sum(kernel_values, axis=0) / len(data)

# Graficar
plt.figure(figsize=(10, 6))

# Contribuciones individuales
for i, xi in enumerate(data):
  plt.plot(x_grid, kernel_values[i, :], label=f"K(x|xi={xi})", linestyle='--')

# Función de densidad de kernel completa
plt.plot(x_grid, density_estimate, label="Estimación de Densidad", linewidth=2)

# Decoración del gráfico
plt.title("Construcción de la Función de Densidad de Kernel")
plt.xlabel("x")
plt.ylabel("Densidad")
plt.legend()
plt.grid(True)
plt.show()
```
